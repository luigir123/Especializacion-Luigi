```python
import random

numero_persona = int(input("Dame un número: "))

# La máquina decide par o impar
eleccion_maquina = random.choice(["par", "impar"])

print("Tu número es:", numero_persona)
print("La máquina eligió:", eleccion_maquina)

# Determinar si el número es par o impar
if numero_persona % 2 == 0:
    resultado = "par"
else:
    resultado = "impar"

print("El número es:", resultado)

# Ver quién gana
if resultado == eleccion_maquina:
    print("¡Ha ganado la máquina, MATRIX!")
else:
    print("¡Ha ganado la persona!")
```

```python
import pandas as pd

ruta_archivo = "/content/compras_semana.csv"

datos = pd.read_csv(ruta_archivo)
datos['total'] = datos['precio'] * datos['cantidad']

total_compras= datos['total'].sum()
print("Total de compras:", total_compras)

promedio_compras = datos['total'].mean()
print("Promedio de compras:", promedio_compras)

mediana_compras = datos['total'].median()
print("Mediana de compras:", mediana_compras)

compra_minima = datos['total'].min()
print("Compra mínima:", compra_minima)

std = datos['total'].std()
print("Desviación estándar:", std)

var = datos['total'].var()
print("Varianza:", var)

resgistro_conteo = datos['total'].count()
print("Registro de conteo:", resgistro_conteo)

datos['total'].describe()
```
